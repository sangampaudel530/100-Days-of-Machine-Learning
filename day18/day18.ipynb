# Web sraping using the general python http requests 

```python
import requests 
from bs4 import BeautifulSoup
```

```python
url="https://www.espn.com/soccer/team/_/id/7635/sport"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                  "(KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.5",
    "Referer": "https://www.google.com/",
    "Connection": "keep-alive"
}

response=requests.get(url,headers=headers)
print(response.status_code)

```

```python
# creating the object of the html parser using the beautifulsoup 
soup=BeautifulSoup(response.text,'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.find_all('h1')[0].text
```

```python
soup.find_all('h2')[1].text
```

```python
for i in soup.find_all('h2'):
    print(i.text)
```

```python
soup.find_all('span', {'class':'Nav__Text'})[12].text.strip()
```

```python
soup.find_all('div',class_='Schedule__Meta__Time ttu fw-medium')
```

```python
for i in soup.find_all('div', class_='ClubhouseHeader__TeamDetails flex items-center mt3'):
    print(i.text)
```
