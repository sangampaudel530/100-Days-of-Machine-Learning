```python
import os, time, shutil

def auto_backup():
    src = "/kaggle/working/runs"
    dst = "/kaggle/working/backup_runs"

    while True:
        if os.path.exists(src):
            shutil.copytree(src, dst, dirs_exist_ok=True)
        time.sleep(300)  # backup every 5 minutes

import threading
threading.Thread(target=auto_backup, daemon=True).start()
```

```python
from ultralytics import YOLO

model = YOLO("yolo26n.pt")

results = model.train(
    data="/kaggle/input/datasets/sangampaudel530/visdrone-finalized-dataset/aerial_object_detection_v1/data.yaml",
    epochs=100,
    imgsz=640,
    batch=-1,
    save=True,
    save_period=5   #  save every 5 epochs
)
```
