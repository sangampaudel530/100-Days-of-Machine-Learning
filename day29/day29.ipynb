## Machine learning pipeline 

```python
import numpy as np 
import pandas as pd 
```

```python
df=pd.read_csv('tested.csv')
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId', 'Name','Ticket','Cabin'],axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.tree import DecisionTreeClassifier
```

```python
#train test split 
X_train, X_test, y_train,y_test=train_test_split(df.drop(columns=['Survived'],axis=1),df['Survived'],test_size=0.2)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_train.isnull().sum()
```

```python
si=SimpleImputer()
```

```python
X_train_age_fare=si.fit_transform(X_train[['Age','Fare']])
X_test_age_fare=si.fit_transform(X_test[['Age','Fare']])
```

```python
X_train_age_fare.shape
```

```python
X_train_age_fare
```

```python
ohe_sex=OneHotEncoder()
ohe_embarked=OneHotEncoder()
```

```python
X_train_sex=ohe_sex.fit_transform(X_train[['Sex']]).toarray()
X_test_sex=ohe_sex.transform(X_test[['Sex']]).toarray()
```

```python
X_train_embarked=ohe_embarked.fit_transform(X_train[['Embarked']]).toarray()
X_test_embarked=ohe_embarked.fit_transform(X_test[['Embarked']]).toarray()
```

```python
X_train_sex.shape, X_train_embarked.shape
```

```python
X_train_rem=X_train.drop(columns=['Age','Sex','Embarked','Fare'])
X_test_rem=X_test.drop(columns=['Age','Sex','Embarked','Fare'])
```

```python
X_train_transformed=np.concatenate((X_train_rem,X_train_age_fare,X_train_sex, X_train_embarked),axis=1)
X_test_transformed=np.concatenate((X_test_rem,X_test_age_fare,X_test_sex, X_test_embarked),axis=1)
```

```python
X_train_transformed
```

```python
X_train_transformed.shape
```

```python
X_test_transformed.shape
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(X_train_transformed,y_train)
```

```python
y_pred=clf.predict(X_test_transformed)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
