# Fetching the data from the api 

```python
import pandas as pd 
import requests
url ='https://api.themoviedb.org/3/movie/popular?api_key=6f1694bd1d4203b31ac7e899e0bf831e&language=en-US&page=1'
response=requests.get(url)
df=pd.DataFrame(response.json()['results'])[['id', 'title', 'popularity', 'release_date', 'overview', 'original_title', 'vote_count', 'vote_average']]
```

```python
df.head()
```

```python
for i in range(1, 10001):
    url = 'https://api.themoviedb.org/3/movie/popular?api_key=6f1694bd1d4203b31ac7e899e0bf831e&language=en-US&page={}'.format(i)
    response = requests.get(url)
    data = response.json()
    if 'results' in data:
        temp_df = pd.DataFrame(data['results'])[['id', 'title', 'popularity', 'release_date', 'overview', 'original_title', 'vote_count', 'vote_average']]
        df = pd.concat([df, temp_df], ignore_index=True)
    else:
        print(f"No results found on page {i}. Stopping.")
        break
```
