## Handling Date and Time 

```python
import pandas as pd 
import numpy as np 
```

## Handling Dates 

```python
df=pd.read_csv('orders.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
df['year']=df['date'].dt.year
```

```python
df.head()
```

```python
df['month_no']=df['date'].dt.month
```

```python
df.head()
```

```python
df['month_name']=df['date'].dt.month_name()
```

```python
df.head()
```

```python
df['day']=df['date'].dt.day
```

```python
df.head()
```

```python
df['day_name']=df['date'].dt.day_name()
```

```python
df.head()
```

```python
df['is_weekend']=np.where(df['day_name'].isin(['Sunday','Saturday']),1,0)
```

```python
df.head()
```

```python
df['quarter']=df['date'].dt.quarter
```

```python
df.head()
```

```python
df['semester']=np.where(df['quarter'].isin([1,2]),1,2)
```

```python
df.head()
```

## Time Elapsed between dates 

```python
import datetime
today=datetime.datetime.today()
today-df['date']
```

```python
(today-df['date']).dt.days
```

```python
np.round((today-df['date']) / np.timedelta64(1, 'W'),1)
```
